```python
data['institute_type'] = [0 if x == 'IIT' else 1 for x in data['institute_type']]
data['institute_type']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
data['quota'] = le.fit_transform(data['quota'])
data['quota'].value_counts()
```

```python
data['pool'] = [0 if x == 'Gender-Neutral' else 1 for x in data['pool']]
data['pool']
```

```python
#college names
data['degree_short'] = le.fit_transform(data['degree_short'])
data['degree_short'].value_counts()
```

```python
data['program_duration'] = [0 if x == '4 Years' else 1 for x in data['program_duration']] 
data['program_duration']
```

```python
data['category']= le.fit_transform(data['category'])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python

x=data.iloc[:,0:10].values
y=data.iloc[:,-1].values
data.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier=RandomForestClassifier(n_estimators=100,criterion="entropy")
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_test,y_pred.round())
cm
```

```python
acc=accuracy_score(y_test,y_pred.round())
acc
```

```python
cr=classification_report(y_test,y_pred)
cr
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators=[int(x) for x in np.linspace(10,80,num=5)]
```

```python
max_features=["auto","sqrt","log2"]
```

```python
max_depth=[int(x) for x in np.linspace(10,100,num=5)]
```

```python
min_samples_split=[2,5,10,15]
```

```python
min_samples_leaf=[1,2,4,6]
```

```python
random_grid={"n_estimators":n_estimators,"max_features":max_features,"max_depth":max_depth,"min_samples_split":min_samples_split,"min_samples_leaf":min_samples_leaf,"criterion":["entropy","gini_index"]}
```

```python
rf=RandomForestClassifier()
rf_randomcv=RandomizedSearchCV(estimator=rf,param_distributions=random_grid,n_iter=100,cv=2,verbose=2,random_state=100,n_jobs=-1)
```

```python
rf_randomcv.fit(x_train,y_train)
```

```python
rf_randomcv.best_params_
```

```python
best_random_grid=rf_randomcv.best_estimator_
```

```python
from sklearn.metrics import accuracy_score
y_pred=best_random_grid.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
